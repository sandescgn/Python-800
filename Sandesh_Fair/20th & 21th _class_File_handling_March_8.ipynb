```python
import csv
students = [
    {'name': "Ram", "address":"ktm", "age":23},
    {'name': "Shyam", "address":"ltr", "age":24},
    {'name': "Sita", "address":"Pkr", "age":21},
    {'name': "Gita", "address":"ktm", "age":22},
]

with open('students2.csv', 'w') as f:
    writer = csv.DictWriter(f, fieldnames=students[0].keys()) # fieldnames=['name', 'address', 'age']
    writer.writeheader()
    writer.writerows(students)
```
